```python
import sqlite3
import os
import csv
os.chdir("E:/EBAC/Material/M42-ConsultasavanzadasenSQLconPython")
```

```python
conn = sqlite3.connect("RH.db")
cur = conn.cursor()
```

```python
with open('recursos_humanos.csv') as f:
    reader = csv.reader(f)
    data = list(reader)
```

```python
# Creación de la tabla donde se importará el archivo CSV
cur.execute("""
    CREATE TABLE Detalle (
        ID INTEGER PRIMARY KEY,
        satisfaction_level REAL,
        last_evaluation REAL,
        number_project INTEGER,
        average_montly_hours INTEGER,
        time_spend_company INTEGER,
        Work_accident INTEGER,
        left INTEGER,
        promotion_last_5years INTEGER,
        sales TEXT,
        salary TEXT
        );
""")
```

```python
for renglon in range(1, len(data)):
    cur.execute("""
        INSERT INTO Detalle (satisfaction_level, last_evaluation, number_project, average_montly_hours, time_spend_company, Work_accident, left, promotion_last_5years, sales, salary)
        VALUES (?,?,?,?,?,?,?,?,?,?)
    """, data[renglon])
```

```python
conn.commit()
```

```python
cur.execute("SELECT sales, salary, satisfaction_level FROM Detalle ORDER BY satisfaction_level DESC")
result = cur.fetchall()
result
```

```python
cur.execute("SELECT salary, number_project, satisfaction_level FROM Detalle ORDER BY number_project, satisfaction_level DESC")
result = cur.fetchall()
result
```

```python
cur.execute("SELECT sales, AVG(last_evaluation) FROM Detalle GROUP BY sales")
result = cur.fetchall()
result
```

```python
cur.execute("SELECT sales, AVG(last_evaluation), salary FROM Detalle GROUP BY sales")
result = cur.fetchall()
result
```

```python
cur.execute("SELECT sales, AVG(Work_accident) FROM Detalle GROUP BY sales HAVING AVG(Work_accident) > 0.15")
result = cur.fetchall()
result
```

```python
cur.execute("SELECT sales, AVG(Work_accident) FROM Detalle GROUP BY sales HAVING AVG(Work_accident) > 200")
result = cur.fetchall()
result
```
