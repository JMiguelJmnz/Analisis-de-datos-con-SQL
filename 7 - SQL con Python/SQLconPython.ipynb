```python
import sqlite3
import os
import csv
os.chdir("E:/EBAC/Material/M41-ImportacionSQLconPython")
```

```python
conn = sqlite3.connect("RH.db")
cur = conn.cursor()
```

```python
with open('recursos_humanos.csv') as f:
    reader = csv.reader(f)
    data = list(reader)
```

```python
data
```

```python
len(data)
```

```python
# Creación de la tabla donde se importará el archivo CSV
cur.execute("""
    CREATE TABLE Detalle (
        ID INTEGER PRIMARY KEY,
        satisfaction_level REAL,
        last_evaluation REAL,
        number_project INTEGER,
        average_montly_hours INTEGER,
        time_spend_company INTEGER,
        Work_accident INTEGER,
        left INTEGER,
        promotion_last_5years INTEGER,
        sales TEXT,
        salary TEXT
        );
""")
```

```python
for renglon in range(1, len(data)):
    cur.execute("""
        INSERT INTO Detalle (satisfaction_level, last_evaluation, number_project, average_montly_hours, time_spend_company, Work_accident, left, promotion_last_5years, sales, salary)
        VALUES (?,?,?,?,?,?,?,?,?,?)
    """, data[renglon])
```

```python
conn.commit()
```

```python
# Satisfaccion promedio de empleados que abandonaron la empresa
cur.execute("SELECT AVG(satisfaction_level) FROM Detalle WHERE left = 1")
result = cur.fetchall()
result
```

```python
# Satisfaccion promedio de empleados que no abandonaron la empresa
cur.execute("SELECT AVG(satisfaction_level) FROM Detalle WHERE left = 0")
result = cur.fetchall()
result
```

Los empleados que deciden quedarse tienen un mayor nivel de satisfaccion promedio (66.68%)

```python
cur.execute("SELECT AVG(average_montly_hours) FROM Detalle WHERE salary = 'low' OR salary = 'medium' GROUP BY salary")
result = cur.fetchall()
result
```

Los empleados con salario LOW trabajaron en promedio 200.99 horas mensuales y los empleados con salario Medium trabajaron en promedio 201.33 horas mensuales

```python
cur.execute("SELECT * FROM Detalle WHERE promotion_last_5years = 1 AND left = 1")
result = cur.fetchall()
result
```

```python
cur.execute("SELECT * FROM Detalle WHERE last_evaluation >= 0.9")
result = cur.fetchall()
result
```
